```python
### find min
from typing import Optional, Any

def find_min(nums: list) -> Any:
    if len(nums) == 0:
        return None
    min = float("inf")
    for num in nums:
        if num < min:
            min = num
    return min

lst = []
lst2 = [100_000_000, 900, 10000, 10, 999999]

print(find_min(lst))
print(find_min(lst2))
```
