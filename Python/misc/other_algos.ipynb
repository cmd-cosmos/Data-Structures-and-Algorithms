```python
# 2 sum

def twoSum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        if target - num in seen:
            return seen[target - num], i
        seen[num] = i
    return num

nums = [2,9,4,2,9,7,8]
twoSum(nums, 13)
```

```python
# kadane's algorithm -> max subarray

def maxSubarray(arr):
    maxSum = float('-inf')
    curr = 0
    for num in arr:
        curr = max(num, curr + num)
        maxSum = max(maxSum, curr)
    return maxSum

nums = [1,7,3,6,4,9,9,8,4,2,1]
print(maxSubarray(nums))

arr = [-2, -3, 4, -1, -2, 1, 5, -3]
print(maxSubarray(arr=arr))
```

```python
# sieve of eratosthenes --> all prime numbers smaller than n

def sieveOfErathosthenes(n):
    prime = [True for i in range(n + 1)]
    p = 2
    while (p * p <= n):
        if (prime[p] == True):
            for i in range(p*p, n+1, p):
                prime[i] = False
        p += 1
    
    for p in range(2, n+1):
        if prime[p]:
            print(p)

sieveOfErathosthenes(15)
```

```python
import random

def shuffle(arr):
    for i in range(len(arr) - 1, 0, -1):
        j = random.randint(0, i)
        arr[i], arr[j] = arr[j], arr[i]
    return arr

data = [1,3,5,7,9,8,0]
shuffle(data)
```
