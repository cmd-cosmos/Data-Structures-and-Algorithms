```python
# odd/even 

def isEven(n):
    return (n & 1) == 0

for i in range(5):
    print(i, isEven(i))
```

```python
# Brian Kernighan's algorithm -> efficient method used to count the number of set bits (1s) in the binary representation of an integer

def countSetBits(n):
    print(f"{n} -> {bin(n)}")
    count = 0
    while n:
        n = n & (n - 1)
        count += 1 
    print(f"set bit count: {count}\n")

countSetBits(20)
countSetBits(100)
countSetBits(1000)
```

```python
# check if power of 2

def isPowerOfTwo(n):
    return n > 0 and (n & (n - 1)) == 0

for n in [1,4,5,8,10,16]:
    print(f"{n:4} | {isPowerOfTwo(n)}")
```

```python
# swapping 2 nums

def swap(a, b):
    a ^= b
    b ^= a
    a ^= b
    return a, b

swap(10, 200)
```
