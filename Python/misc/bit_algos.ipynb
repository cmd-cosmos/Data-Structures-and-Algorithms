```python
# odd/even 

def isEven(n):
    return (n & 1) == 0

for i in range(5):
    print(i, isEven(i))
```

```python
# Brian Kernighan's algorithm -> efficient method used to count the number of set bits (1s) in the binary representation of an integer

def countSetBits(n):
    print(f"{n} -> {bin(n)}")
    count = 0
    while n:
        n = n & (n - 1)
        count += 1 
    print(f"set bit count: {count}\n")

countSetBits(20)
countSetBits(100)
countSetBits(1000)
```

```python
# check if power of 2

def isPowerOfTwo(n):
    return n > 0 and (n & (n - 1)) == 0

for n in [1,4,5,8,10,16]:
    print(f"{n:4} | {isPowerOfTwo(n)}")
```

```python
# swapping 2 nums

def swap(a, b):
    a ^= b
    b ^= a
    a ^= b
    return a, b

swap(10, 200)
```

```python
def findUnique(nums):
    res = 0
    for num in nums:
        res ^= num
    return res

nums = [1,4,2,1,2]
findUnique(nums=nums)
```

```python
# single bit manips

def getBit(n, i):
    return (n >> i) & 1

def setBit(n, i):
    return n | (1 << i)

def clearBit(n, i):
    return n & ~(1 << i)

def toggleBit(n, i):
    return n ^ (1 << i)

n = 0b1010
print(bin(getBit(n, 1)))
print(bin(setBit(n, 0)))
print(bin(clearBit(n, 3)))
print(bin(toggleBit(n, 2)))
```

```python
# reverising bits

def reverseBits(n, bits=8):
    res = 0
    for _ in range(bits):
        res = (res << 1) | (n & 1)
        n >>= 1
    return res

n = 0b110101
bin(reverseBits(n, 6))
```

```python
# finding missing num within a range

def findMissing(arr, n):
    xor_all = 0
    xor_arr = 0
    for i in range(1, n + 1):
        xor_all ^= i
    for num in arr:
        xor_arr ^= num
    return xor_all  ^ xor_arr

# 3 is missing in the sequence so the sequence is broken 
arr = [1,2,4,5,6]
findMissing(arr, 6)
```
