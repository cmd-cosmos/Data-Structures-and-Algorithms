```python
### basic bst implementation and dfs traversal

class Node:
    def __init__(self, data) -> None:
        self.data  = data
        self.left  = None
        self.right = None
```

```python
# type: ignore

class BST:
    def __init__(self) -> None:
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data=data)
        else:
            self._insert(self.root, data)

    def _insert(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data=data)
            else:
                self._insert(node.left, data=data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data=data)
            else:
                self._insert(node.right, data=data)
            
    ### dfs --> pre, in, post order traversal

    def preorder(self):
        res = []

        def dfs(node):
            if not node:
                return
            res.append(node.data)
            dfs(node.left)
            dfs(node.right)
        
        dfs(self.root)
        return res
        
```

```python
bst = BST()

for val in [50, 30, 70, 20, 40, 60, 80]:
    bst.insert(val)

bst.preorder()
```
