```python
# pattern matching

def naive_search(txt, pattern):
    """
    return start to end index of the pattern in the string.
    exclusive of the upper bound
    """
    n,m = len(txt), len(pattern)
    for i in range(n-m+1):
        if txt[i:i+m] == pattern:
            return i, i+m
    return -1

naive_search("manbattab", "tta")
```

```python
# check palindromes

def is_palindrome(string):
    return string == string[::-1]

print(is_palindrome("racecar"))
print(is_palindrome("pop"))
print(is_palindrome("batman"))
print(is_palindrome("bob"))
```
