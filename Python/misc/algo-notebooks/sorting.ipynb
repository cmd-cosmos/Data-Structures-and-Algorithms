```python
# bubble sort

def bubbleSort(arr):
    print("\nbubble sort: ")
    n = len(arr)
    for i in range(n):
        for j in range(n - i -1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                print(arr)
    print("\nsorted array:", arr)
    return arr

arr1= [1,9,8,5,7,2,1,3,0,9,6,5,2]
print(f"arr1 = {arr1}")
arr1Sorted = bubbleSort(arr=arr1)
```

```python
# insertion sort

def insertionSort(arr):
    print("\ninsertion sort")
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
            print(arr)
        arr[j + 1] = key
    print(f"\nsorted arr: {arr}")
    return arr

array = [1,9,8,5,7,2,1,3,0,9,6,5,2]
print(f"unsorted arr: {array}")
sortedArray = insertionSort(arr=array)
```

```python
# quick sort

def quickSort(arr):
    n = len(arr)

    if n <= 1:
        return arr
    
    pivot = arr[n // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    print(arr)
    return quickSort(left) + middle + quickSort(right)

array2 = [1,9,8,5,7,2,1,3,0,9,6,5,2]
print(f"unsorted arr: {array2}\n")
quickSort(arr=array2)
```

```python
# merge sort

def mergeSort(arr):
    n = len(arr)
    if n <= 1:
        return arr
    
    mid = n // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    print(arr)
    return merge(left, right)

def merge(left, right):
    res = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    
    res.extend(left[i:])
    res.extend(right[j:])
    
    return res

array3 = [1,9,8,5,7,2,1,3,0,9,6,5,2]
print(f"unsorted arr: {array3}\n")
mergeSort(array3)
```
