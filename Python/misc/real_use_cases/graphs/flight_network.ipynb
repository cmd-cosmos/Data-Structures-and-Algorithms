```python
import heapq

class FlightNetwork:
    def __init__(self) -> None:
        self.graph = {}
    
    def add_route(self, from_airport, to_airport, dist):
        self.graph.setdefault(from_airport, []).append((to_airport, dist))
        self.graph.setdefault(to_airport, []).append((from_airport, dist))
    
    def shortest_path(self, start, end):
        pq = [(0, start, [])]
        visited = set()

        while pq:
            cost, node, path = heapq.heappop(pq)

            if node in visited:
                continue
            visited.add(node)
            path = path + [node]

            if node == end:
                return cost, path
            
            for neighbor, weight in self.graph.get(node, []):
                if neighbor not in visited:
                    heapq.heappush(pq, (cost + weight, neighbor, path))
        
        return float("inf"), []
```

```python
network = FlightNetwork()

network.add_route("A", "B", 360)
network.add_route("C", "D", 1000)
network.add_route("A", "E", 120)
network.add_route("E", "D", 840)

cost, route = network.shortest_path("A", "D")
print(f"shortest path: {route}")
print(f"total distance: {cost}")
```
