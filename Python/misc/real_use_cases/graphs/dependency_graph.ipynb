```python
# fault propogation in a system --> sensor failure propogation through a dependency graph
from collections import defaultdict

class SubSysGraph:
    def __init__(self) -> None:
        self.graph = defaultdict(list)
    
    def add_dependency(self, subsys, depends_on):
        self.graph[subsys].append(depends_on)

    def dependencies_of(self, subsys):
        return self.graph[subsys]
```

```python
gnc = SubSysGraph()

gnc.add_dependency("Guidance", "IMU")
gnc.add_dependency("Guidance", "GPS")
gnc.add_dependency("Control", "Guidance")
gnc.add_dependency("Control", "Actuators")
gnc.add_dependency("Telemetry", "Guidance")
```

```python
# type:ignore
def prop_failure(graph, failed_node, visited=None):
    if visited is None:
        visited = set()

    visited.add(failed_node)

    for subsys, deps in graph.graph.items():
        if failed_node in deps and subsys not in visited:
            prop_failure(graph=graph, failed_node=subsys, visited=visited)

    return visited
```
