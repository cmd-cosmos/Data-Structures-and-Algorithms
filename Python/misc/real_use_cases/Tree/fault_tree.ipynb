```python
class FaultTree:
    def __init__(self, name) -> None:
        self.name = name
        self.children = []

    def add(self, *nodes):
        self.children.extend(nodes)

    def __str__(self) -> str:
        return f"FaultTree Member -> {self.name}"
```

```python
# tree root
pressure_loss_root = FaultTree("Loss of Hydraulic Pressure")

print(pressure_loss_root)
```
