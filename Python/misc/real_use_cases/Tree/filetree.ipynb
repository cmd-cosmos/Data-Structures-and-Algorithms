```python
class File:
    def __init__(self, name, size=0, is_dir=False) -> None:
        self.name = name
        self.size = size
        self.is_dir = is_dir
        self.children = []
        self.parent = None
```

```python
class FileTree:
    def __init__(self, root=None) -> None:
        self.root = File(name="root", is_dir=True)

    def add_file(self, parent_node, new_file):
        if not parent_node.is_dir:
            raise Exception("Cannot add file to file")
        
        new_file.parent = parent_node
        parent_node.children.append(new_file)

    def walk(self, node=None, level=0):
        if node is None:
            node = self.root
        print(" " * level * 2 + node.name)
        if node.is_dir:
            for child in node.children:
                self.walk(child, level+1)
```
