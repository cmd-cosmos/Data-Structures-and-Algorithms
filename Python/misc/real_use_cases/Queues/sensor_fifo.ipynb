```python
from collections import deque
import random
import time
```

```python
class SensorFIFO:
    def __init__(self, capacity) -> None:
        self.capacity = capacity
        self.buffer   = deque(maxlen=capacity)
    
    def add_reading(self, val):
        self.buffer.append(val)
        print(f"Added: {val:5} | Data Buffer: {list(self.buffer)}")

    def get_latest(self):
        if self.buffer:
            return self.buffer[-1]
        return None

    def get_all_readings(self):
        return list(self.buffer)

```

```python
fifo_sensors = SensorFIFO(capacity=5)

for _ in range(5):
    sample_readings = round(random.uniform(10, 20), 2) # round a random val b/w 10-20 to 2 decimals
    fifo_sensors.add_reading(sample_readings)
    time.sleep(0.5)
```

```python
fifo_sensors.get_all_readings()
```
