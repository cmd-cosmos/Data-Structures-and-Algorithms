```python
# bubble sort

def bubbleSort(arr):
    print("\nbubble sort: ")
    n = len(arr)
    for i in range(n):
        for j in range(n - i -1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                print(arr)
    print("\nsorted array:", arr)
    return arr

arr1= [1,9,8,5,7,2,1,3,0,9,6,5,2]
print(f"arr1 = {arr1}")
arr1Sorted = bubbleSort(arr=arr1)
```
