```python
from typing import Optional, Any

class Node:
    def __init__(self, data: Any, left_child:Optional['Node']=None, right_child:Optional['Node']=None) -> None:
        self.data: Any = data
        self.left_child:  Optional['Node'] = left_child
        self.right_child: Optional['Node'] = right_child
```

```python
'''
     1
  2      3
4   5
'''

root = Node(1, Node(2, Node(4), Node(5)), Node(3))
```

```python
# preorder traversal

def preorder(node: Optional['Node']) -> None:
    if node is None:
        return
    print(node.data)
    preorder(node=node.left_child)
    preorder(node=node.right_child)

preorder(root)
```

```python
# inorder traversal

def inorder(node:Optional['Node']) -> None:
    if node is None:
        return
    inorder(node=node.left_child)
    print(node.data)
    inorder(node=node.right_child)

inorder(root)
```
