```python
#type: ignore
from typing import Any, Optional

class Node:
    def __init__(self, data: Any) -> None:
        self.data: Any = data
        self.next: Optional['Node'] = None
        self.prev: Optional['Node'] = None
```

```python
class DoublyLinkedList:
    def __init__(self) -> None:
        self.head: Optional['Node'] = None
        self.contents: list[Any]    = []
        self.__size: int          = 0

    def isEmpty(self) -> bool:
        return not self.__size
    
    def insertNode(self, data: Any):
        newNode: Node = Node(data=data)

        if self.head:
            curr = self.head
            while curr.next:
                curr = curr.next
            temp = curr
            curr.next = newNode
            newNode.prev = temp 
            self.__size += 1
            self.contents.append(data)
        else:
            self.head = newNode
            self.__size += 1
            self.contents.append(data)
            return
        
    def printList(self):
        curr = self.head
        print("NULL <--> ", end="")
        while curr:
            print(f"{curr.data} <--> ", end="")
            curr = curr.next
        print("NULL")
            
```

```python
dll = DoublyLinkedList()
```

```python
dll.isEmpty()
```

```python
dll.insertNode(10)
dll.insertNode(20)
dll.insertNode(30)
dll.insertNode(40)
```
