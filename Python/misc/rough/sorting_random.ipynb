```python
def bubble_sort(arr):
    n = len(arr)
    iter = 0
    for i in range(n-1):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                iter += 1
                print(iter, arr)
    return arr

arr = [36, 22, 98, -121, 11, 2, 13, 1, 34, -10, -4, 170, 180, 200, 220]
print(arr, end="\n\n")
bubble_sort(arr)
```

```python
def opt_bubble_srt(arr):
    n = len(arr)
    iter = 0
    for i in range(n-1):
        swap = False
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                iter += 1
                swap = True
                print(iter, arr) 
        if not swap:
            break
    return arr

arr2 = [9, 7, 10, 1, 80, 90, 100, 110]
print(arr2, end="\n\n")
opt_bubble_srt(arr2)
```
