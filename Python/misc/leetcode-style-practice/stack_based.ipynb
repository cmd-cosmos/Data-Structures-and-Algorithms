```python
### valid parenthesis problem -> are open brackets closed

def isValid(string):
    stack = []
    hashmap = {
        ")" : "(",
        "]" : "[",
        "}" : "{"
    }
    for char in string:
        if char in hashmap:
            if not stack or stack.pop() != hashmap[char]:
                return False
        else:
            stack.append(char)
            print(stack)
    return not stack

case1 = r"()[]{}"
case2 = r"([]}"
print(isValid(case1))
print(isValid(case2))
```
