```python
### valid parenthesis problem -> are open brackets closed

def isValid(string):
    stack = []
    hashmap = {
        ")" : "(",
        "]" : "[",
        "}" : "{"
    }
    for char in string:
        if char in hashmap:
            if not stack or stack.pop() != hashmap[char]:
                return False
        else:
            stack.append(char)
            print(stack)

    return not stack

case1 = r"()[]{}"
case2 = r"([)]{}"
print(isValid(case1))
print(isValid(case2))
```

```python
class MinStack:
    def __init__(self) -> None:
        self.stack = []
        self.min_stack = []

    def push(self, val):
        self.stack.append(val)
        if not self.min_stack or val < self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        if self.stack.pop() == self.min_stack[-1]:
            self.min_stack.pop()
        
    def top(self):
        return self.stack[-1]
    
    def getMin(self):
        return self.min_stack[-1]
    
    def printStack(self):
        print(f"Stack: {self.stack}")
        print(f"Min Stack: {self.min_stack}")
    
ms = MinStack()

ms.push(1)
ms.push(5)
ms.push(2)
ms.push(3)
ms.push(1)

ms.printStack()

ms.pop()
ms.printStack()
```
