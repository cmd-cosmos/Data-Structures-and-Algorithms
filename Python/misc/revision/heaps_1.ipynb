```python
# purpose of the heap is to get smallest/largest val quickly
# basically a mod of the binary tree

# min heap -> parent <= left_node
#          -> parent <= right_node
```

```python
import heapq

h = []

heapq.heappush(h, 10)
heapq.heappush(h, 4)
heapq.heappush(h, 1)

print(h)
```

```python
smallest = heapq.heappop(h)
print(smallest)
print(h)
```

```python
### heapify converts a list into a heap
### min heap
### heap guarantees smallest on top and not a sorted list

nums = [10, 6,2,9,4,1]

heapq.heapify(nums)
print(nums)
```
